```python
from sklearn.datasets import make_classification
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import pandas as pd
```

```python
# Generate and dataset for Logistic Regression
x, y = make_classification(
    n_samples=100,
    n_features=1,
    n_classes=2,
    n_clusters_per_class=1,
    flip_y=0.03,
    n_informative=1,
    n_redundant=0,
    n_repeated=0
)
```

```python
plt.scatter(x, y, c=y, cmap='rainbow')
plt.title('Scatter Plot of Logistic Regression')
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=1)
```

```python
#log_reg = LogisticRegression()
log_reg = svm.SVC()
log_reg.fit(x_train, y_train)
print(log_reg)
```

```python
log_reg = svm.SVC(kernel='linear')
log_reg.fit(x_train, y_train)

# Print parameters using get_params()
params = log_reg.get_params()
print("SVM(", end="")
for key, value in params.items():
  print(f"{key}={value}, ", end="")
print(")")
```

```python
# Show to Coeficient and Intercept
print(log_reg.coef_)
print(log_reg.intercept_)
```

```python
# Perform prediction using the test dataset
y_pred = log_reg.predict(x_test)
```

```python
# Show the Confusion Matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Create an SVC object with probability estimates enabled
log_reg = SVC(kernel='linear', probability=True)

# Fit the model (replace with your actual training data)
log_reg.fit(x_train, y_train) # This line was commented out

# Get decision function values (replace x_test with your test data)
y_scores = log_reg.decision_function(x_test)

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_scores)

# Calculate AUC
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

# Assuming you have x_train, y_train, x_test, y_test from your code
# Create and train the SVM model
svm_model = SVC(kernel='linear')
svm_model.fit(x_train, y_train)

# Make predictions
y_pred = svm_model.predict(x_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
